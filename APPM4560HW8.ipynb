```python
import numpy as np
import matplotlib.pyplot as plt

N = [100, 1000, 10000, 100000]
t_max = 5
dt = 0.01
times = np.arange(0, t_max, dt)
sd = [1/3, 2/3, 0, 0]

def y(t):
    y = 1/4 + (1/12)*np.exp(-2*t) + (1/6)*np.exp(-t)*(np.cos(t) + 2*np.sin(t))
    return y


for n in N:

    fraction = []
    states = np.random.choice([1, 2, 3, 4], size=n, p=sd )

    for t in times:
        fraction.append(np.mean(states == 1))

        jumps = np.random.random(n) < dt
        states[jumps] = (states[jumps] % 4) + 1

    plt.plot(times, fraction, label=f'N={n}')


plt.plot(times, y(times), label='y1(t)')

plt.xlabel('t')
plt.ylabel('Fraction in State 1')
plt.title('Sim vs Theory in State 1')
plt.legend()
plt.xlim([0, 5])
plt.ylim([0, 0.5])
plt.show()
```
