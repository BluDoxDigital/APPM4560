```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt

## HW 1 QUESTION 3 ##
'''
Estimate the integral by estimating the area under the function f . To do this, sample N points uniformly in
the square [0, 1]2 and calculate the fraction of those that are below the function f . By doing this, you will
obtain an estimate E(N ). Plot E(N ) as a function of N on a plot where the horizontal axis is logarithmic.
Take N = floor(10x) with x = 1, 1.1, 1.2, . . . , 4.9, 5.
'''

X = np.linspace( 1,5, 41 )
N = [ math.floor( 10**x ) for x in X ]
E = []

for n in N:
  u = np.random.rand(n)
  v = np.random.rand(n)
  w = [ j**2/(j**6+1) for j in u ]
  s = 0
  for c in range(0, n):
    if v[c] <= w[c] :
      s+=1
  e = s/n
  E.append(e)

x = [ math.log10(n) for n in N ]
y = E

plt.axhline( y=math.pi/12, color="red", label="pi/12" )

plt.plot( x, y, label="E[N]" )
plt.legend()

plt.show()


print( sum(E)/40 )
print( math.pi/12 )
```
